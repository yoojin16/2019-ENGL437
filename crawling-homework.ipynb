## 과제: 네이버 베스트셀러 주간 순위 중 11번가 부분 책의 정보를 가져오기
첫페이지에 있는 25권의 책목록에서 링크를 추출하여 각 추출된 링크로 접속하여 책이름, 저자, 가격 정보를 출력

```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://book.naver.com/bestsell/bestseller_list.nhn?cp=morning365')
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

```python
for n in range(0,25):
    dt = bsObject.find("dt", {"id":"book_title_"+str(n)})
    title = dt.select('a')[0].get_text()
    author = dt.next_sibling.next_sibling.select('a')[0].get_text()
    publisher = dt.next_sibling.next_sibling.get_text().split()[-1]
    price = bsObject.find_all("del",{"class":"price_off"})[n].get_text()
    print(n+1, title, author, publisher, price)
```

```python
# Create a file to write to, add headers row
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'author', 'publisher', 'price'])
        
    for n in range(0,25):
        dt = bsObject.find("dt", {"id":"book_title_"+str(n)})
        title = dt.select('a')[0].get_text()
        author = dt.next_sibling.next_sibling.select('a')[0].get_text()
        publisher = dt.next_sibling.next_sibling.get_text().split()[-1]
        price = bsObject.find_all("del",{"class":"price_off"})[n].get_text()
        writer.writerow([n+1, title, author, publisher, price])
```
