```python
# TensorFLow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

### data preparation

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

### create model

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

### train model

```python
model.fit(x_train, y_train, epochs=1000)

# x_train에는 13개의 column, y_train에는 1개의 column이 들어갈 것.
# fit은 훈련의 의미. 데이터의 80%정도를 이용
```

### evaluate model

```python
model.evaluate(x_test, y_test)

# 테스트하는 과정. test용 데이터를 사용.
```

### predict model

```python
model.predict(np.array([0,1,50,1,0,5,.5,10,1000,100, 10,100,1]).reshape(1,13))
# predict는 실제 상용화할 때 사용. 
```
