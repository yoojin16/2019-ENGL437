```python
# TensorFLow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

### data preparation

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

### create model

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

### train model

```python
model.fit(x_train, y_train, epochs=1000)

# x_train 13 column, y_train 1 column  .
# fit  .  80% 
```

### evaluate model

```python
model.evaluate(x_test, y_test)

#  . test  .
```

### predict model

```python
model.predict(np.array([0,1,50,1,0,5,.5,10,1000,100, 10,100,1]).reshape(1,13))
# predict    . 
```
